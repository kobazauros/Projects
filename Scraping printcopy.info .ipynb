#### Import of libraries, prepare initial request of brands with USLs from printcopy.info

```python
import requests
r = requests.get('https://printcopy.info/?mod=pc')
main_link = 'https://printcopy.info/'
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
import re
brand = soup.find_all('a', attrs={'href':re.compile('brand')})
```

```python
brand_dict = {k.text:k['href'] for k in brand}
```

```python
brand_dict
```

#### Secondary request to the site printcopy.info, allows you to get a list of models with URLs

```python
model_dict_all = dict()
for i in brand_dict:
    page = 1    
    brand_r = requests.get('https://printcopy.info/'+ brand_dict[i])
    while brand_r.ok:
        brand_r = requests.get('https://printcopy.info/'+ brand_dict[i]+'&page='+str(page))        
        soup_model = BeautifulSoup(brand_r.text, 'html.parser')
        model = soup_model.find_all('a', attrs={'href':re.compile('model')})
        model_dict = {k.text:k['href'] for k in model}        
        model_dict_all.update(model_dict)
        page = page + 1
```

#### getting a list of free proxies

```python
import requests
import random
from bs4 import BeautifulSoup as bs
def get_free_proxies():
    url = "https://free-proxy-list.net/"
    # посылаем HTTP запрос и создаем soup объект
    soup = bs(requests.get(url).content, "html.parser")
    proxies = []
    for row in soup.find("table", attrs={"class": "table table-striped table-bordered"}).find_all("tr")[1:]:
        tds = row.find_all("td")
        try:
            ip = tds[0].text.strip()
            port = tds[1].text.strip()
            host = f"{ip}:{port}"
            proxies.append(host)
        except IndexError:
            continue
    return proxies
```

```python
proxies = get_free_proxies()
```
